заполним `kitchen_area` следующим образом:
рассчитаем `non_living_area` как разницу между общей площадью и жилой
посчитаем среднее значение отношения площади кухни к нежилой площади
заполним произведением нежилой площади на это рассчитанное значение

```python
df['kitchen_to_non'] = df.kitchen_area / df.non_living_area
kitchen_to_non_mean = df.kitchen_to_non.mean()
kitchen_to_non_mean
df.loc[df.kitchen_area.isna(), 'kitchen_area'] = df[df.kitchen_area.isna()]['non_living_area'] * kitchen_to_non_mean
```

рассчитаем цену метра недвижимости в населенных пунктах с наибольшим количеством объявлений

```python
popular_localities = df.groupby(by='locality_name')['last_price'].count().sort_values(ascending=False)[:10]
popular_localities = list(popular_localities.index)
# popular_localities
df_in_popular = df[df.locality_name.isin(popular_localities)]
avg_price = df_in_popular.pivot_table(index='locality_name', values=['total_area', 'last_price'], aggfunc='sum')
avg_price['price_per_meter'] = (avg_price.last_price / avg_price.total_area).round()
avg_price.sort_values(by='price_per_meter', ascending=False)
```

Если выбрать 10 населенных пунктов с наибольшим количеством объявлений, видна значительная разница в стоимости недвижимости между населенными пунктами. Ожидаемо самым дорогим является Санкт-Петербург, самым доступным - Выборг, разница в стоимости квадратного метра - почти в 2 раза
